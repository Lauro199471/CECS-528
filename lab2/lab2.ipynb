```python
import numpy as np
import matplotlib.pyplot as plt  
import math 
```

```python
x = np.arange(.001,5,.01);
```

```python
fx = (x**2) + 1;

plt.plot(x,fx,"green");
plt.axhline(y=0, color='black', linestyle='-' )
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();
plt.title('f(x) = (x^2) + 1');
plt.savefig('f(x)');
```

```python
gx = x + 8;

plt.plot(x,gx,"blue");
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();
plt.title('g(x) = x + 8');
plt.savefig('g(x)');
```

```python
plt.plot(x,fx,'green',x,gx,"blue");
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();

plt.title('Show f(x)(green) + g(x)(blue) = Î˜(max(f(x),g(x)))');
plt.savefig('f(x)&g(x)');
```

```python
maxList = [];
for i in range(0 , x.size):
    if fx[i] > gx[i]:
        maxList.append(fx[i]);
    
    elif gx[i] > fx[i]:
        maxList.append(gx[i])
    
    elif gx[i] == fx[i]:
        maxList.append(gx[i])

plt.plot(x,maxList,"red" , linewidth = 8);
plt.plot(x,fx,'green',x,gx,"blue");
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();
plt.title('f(x)(green) + g(x)(blue) = Î˜(max(f(x),g(x))(red))');
plt.savefig('f(x)Cg(x)1');
plt.show();
plt.close();
```

```python
c =  1;
n0 = 0;

plt.plot(x,np.multiply(maxList,c),"red");
plt.plot(x,fx+gx,'purple');

plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();
plt.title('Big â„¦: (f(x) + g(x))(purple) >= C(max(f(x),g(x))(red)), where C = 1 , n>=0');
plt.savefig('f(x)Cg(x)2');
plt.show();
plt.close();
```
