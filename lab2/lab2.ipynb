```python
import numpy as np
import matplotlib.pyplot as plt  
import math 
```

```python
x1 = np.arange(.01,3,0.1);
y1 = (x1*np.log2(x1));

plt.plot(x1,y1,"green");
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();
plt.title('f(x) = x * log(x)');
plt.savefig('f(x)');
```

```python
x2 = np.arange(.01,3,0.1);
y2 = x2 + (x2 * np.log2(x2**2));

plt.plot(x2,y2,"blue");
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();
plt.title('g(x) = x + (x*log(x^2))');
plt.savefig('g(x)');
```

```python
plt.plot(x1,y1,'green',x2,y2,"blue");
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();

plt.title('Show f(x)(green)<=C(g(x)(blue))');
plt.savefig('f(x)&g(x)');
```

```python
c = 2;
n0 = 1;

x3 = np.arange(.01,3,0.01);
y3 = c * (x3 + (x3 * np.log2(x3**2)) );

plt.plot(x1,y1,'green',x3,y3,"blue");

plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.axvline(x=n0, color='red', linestyle='-')

plt.grid();

plt.title('f(x) <= C*g(x), where C =' + str(c) + ', n>=' + str(n0));
plt.savefig('f(x)Cg(x)1');
plt.close();

```
