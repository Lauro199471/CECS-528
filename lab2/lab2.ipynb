```python
import numpy as np
import matplotlib.pyplot as plt  
import math 
```

```python
x1 = np.arange(.001,5,.01);
y1 = (x1**2) + x1 + 7;

plt.plot(x1,y1,"green");
plt.axhline(y=0, color='black', linestyle='-' )
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();
plt.title('f(x) = (x^2) + x + 7');
plt.savefig('f(x)');
```

```python
x2 = np.arange(.001,5,.01);
y2 = x2**2;

plt.plot(x2,y2,"blue");
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();
plt.title('g(x) = x^2');
plt.savefig('g(x)');
```

```python
plt.plot(x1,y1,'green',x2,y2,"blue");
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();

plt.title('Show g(x)(blue)>=C(f(x)(green))');
plt.savefig('f(x)&g(x)');
```

```python
c =  2;
n0 = 4;

x3 = np.arange(.001,5,.01);
y3 = c * (x3**2);

plt.plot(x1,y1,'green',x3,y3,"blue",  linewidth = 2.5);

plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.axvline(x=n0, color='red', linestyle='-')

plt.grid();

plt.title('g(x)(blue) >= C*f(x)(green), where C =' + str(c) + ', n>=' + str(n0));
plt.savefig('f(x)Cg(x)1');
plt.show();
plt.close();

```
