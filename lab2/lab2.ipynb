```python
import numpy as np
import matplotlib.pyplot as plt  
import math 
```

```python
x = np.arange(.001,5,.01);
```

```python
fx =  x * np.log2(x);

plt.plot(x,fx,"green");
plt.axhline(y=0, color='black', linestyle='-' )
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();
plt.title('f(x) = xlogx');
plt.savefig('f(x)');
```

```python
gx = x + (2*x*np.log2(x));

plt.plot(x,gx,"blue");
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();
plt.title('g(x) = x + xlogx^2');
plt.savefig('g(x)');
```

```python
plt.plot(x,fx,'green',x,gx,"blue");
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();

plt.title('Show f(x)(green) = O(g(x)(blue))');
plt.savefig('f(x)&g(x)');
```

```python
maxList = [];
for i in range(0 , x.size):
    if fx[i] > gx[i]:
        maxList.append(fx[i]);
    
    elif gx[i] > fx[i]:
        maxList.append(gx[i])
    
    elif gx[i] == fx[i]:
        maxList.append(gx[i])

plt.plot(x,maxList,"red" , linewidth = 8);
plt.plot(x,fx,'green',x,gx,"blue");
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.grid();
plt.title('f(x)(green) + g(x)(blue) = (max(f(x),g(x))(red))');
plt.savefig('f(x)Cg(x)1');
plt.show();
plt.close();
```

```python
c =  1;
n0 = 1;

plt.plot(x,np.multiply(c , gx),'blue',linestyle='-');
plt.plot(x,fx,'green');

plt.axvline(x=n0, color='red', linestyle='-');
plt.axhline(y=0, color='black', linestyle='-');
plt.axvline(x=0, color='black', linestyle='-');


plt.grid();
plt.title('Big O: f(x)(green) <= Cg(x))(blue) , where C = 1 , n>=1');
plt.savefig('f(x)Cg(x)2');
plt.show();
plt.close();
```
